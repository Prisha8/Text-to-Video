```python
from moviepy.editor import VideoFileClip, AudioFileClip, CompositeAudioClip

# Use paths within Kaggle's working directory
inputvideo = ""
inputaudio = ""
outputvideo = ""

# Load video and audio clips
videoclip = VideoFileClip(inputvideo)
original_audio = videoclip.audio  # Extract original audio
background_audio = AudioFileClip(inputaudio)  # Load new audio

# Combine both audio sources
final_audio = CompositeAudioClip([original_audio, background_audio])

# Set the new mixed audio to the video
finalclip = videoclip.set_audio(final_audio)

# Save the output file
finalclip.write_videofile(outputvideo, fps=30, codec="libx264")

print("Video processing complete. Check 'output4.mp4' in Kaggle's working directory.")
```
